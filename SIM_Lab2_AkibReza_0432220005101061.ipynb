```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assignment 1

m1 = np.random.randint(1,100, size=(3,3))
m2 = np.random.randint(1,100, size=(3,3))

sum_matrix = m1 + m2
sub_matrix = m1 - m2
prod_matrix = np.dot(m1,m2)

print("m1:\n",m1)
print("m2:\n",m2)
print("Summation of m1 and m2\n", sum_matrix)
print("Subtraction of m1 and m2\n",sub_matrix)
print("Product of m1 and m2\n", prod_matrix)


det_m1 = np.linalg.det(m1)
inv_m1 = np.linalg.inv(m1)
rank_m1 = np.linalg.matrix_rank(m1)

print("Determinant of m1: \n", det_m1)
print("Inverse of m1:  \n", inv_m1)
print("Rank of m1: \n", rank_m1)


# Assignment 2

vec1 = np.random.rand(15)
vec2 = np.random.rand(15)
index = np.arange(len(vec1))


print("Vector 1: \n", vec1)
print("Vector 2: \n", vec2)


plt.plot(vec1,index, marker='o', linestyle='-',label="Vector 1")
plt.plot(vec2,index, marker='s', linestyle='--',label="Vector 2")
plt.title("Scatter plot of two vectors")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend("")
plt.grid("true")
plt.show()


MX = np.random.rand(4, 4)
print("4x4 Matrix: \n", MX)

sns.heatmap(MX, annot=True, cmap="Reds",linewidths=0.5,linecolor="black")
plt.title("Heatmap of Matrix")
plt.show()



M_1 = np.random.randint(1,100, size=(4,4))
M_2 = np.random.randint(1,100, size=(4,4))

M_PLUS = M_1 + M_2
M_MINUS = M_1 - M_2
M_PROD = np.dot(M_1,M_2)

print("Matrix 1: \n", M_1)
print("Matrix 2: \n", M_2)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# Sum
axes[0].bar(range(16), M_PLUS.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')

# Difference
axes[1].bar(range(16), M_MINUS.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')

# Product
axes[2].bar(range(16), M_PROD.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
